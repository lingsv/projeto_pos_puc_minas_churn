```python
# Função para contar caracteres especiais em uma string
def contar_caracteres_especiais(texto):
    if pd.isnull(texto):
        return 0
    # Expressão regular atualizada para considerar acentos e cedilha
    return len(re.findall(r'[^a-zA-ZÀ-ÿ\s]', texto))
```

```python
# Contar caracteres especiais nas colunas 'nome' e 'sobrenome'
caracteres_especiais_nome = dataset['nome'].apply(contar_caracteres_especiais).value_counts()
caracteres_especiais_sobrenome = dataset['sobrenome'].apply(contar_caracteres_especiais).value_counts()
print(caracteres_especiais_nome)
print(caracteres_especiais_sobrenome)
```
